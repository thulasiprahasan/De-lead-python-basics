## pip package

`It is the package manager used to install python modules in your system irrespective of the operation system.`

**Steps for windows:**

1. Download `get-pip.py` file available in (https://bootstrap.pypa.io/get-pip.py)
   Save this file in the same folder where you have installed the python3.   
   OR   
   run the following command in cmd:   
   `curl https://bootstrap.pypa.io/get-pip.py -o get-pip.py`
   
   
2. Then execute the following command in your terminal/cmd:     
   `python get-pip.py` (same for mac/windows/linux)

3. Once it got installed verify using the command `pip help` in the cmd
   ![](images/pip-1.jpg)
   
-----


**Steps for ubuntu:**   

1. Open terminal

2. Type the following command and press enter:
   `sudo apt install python3-pip`
   
3. To verify the installation, run the following command in terminal:
   `pip3 --version`
   You will get some valid version number of pip

-------


**Steps for MacOS:**  

1. Download `get-pip.py` file available in (https://bootstrap.pypa.io/get-pip.py)              
   OR   
   run the following command in cmd:   
   `curl https://bootstrap.pypa.io/get-pip.py -o get-pip.py`
   
   
2. Then execute the following command in your terminal/cmd:     
   `python get-pip.py` (same for mac/windows/linux)
