### Functions - Arguments

Function can take arguments 
- Arguments are passed inside the paranthesis of the function
- Arguments make function more useful
- Some functions require arguments
- Inside the function, the arguments are assigned to variables called parameters. 
- Once we define a function, we can reuse the function over and over throughout our program.

##  Arguments Types
1. **Positional arguments** - should be passed first
1. **Keyword arguments** - No order , keyword is matched
1. **Arbitary arguments** - \* \<name\>

# 1. Position arguments 

- Also known as required arguments.


- Position arguments are the arguments passed to a function in correct positional order.


- One-to-One mapping between arguments and parameters with respect to the order they appear.


- No.of arguments in the function call should match exactly with the function definition.

````python
def n(a):
    a=a+1
    print(a)
n(15) # a becomes 15
n(20) # a becomes 20
````

````python
def cool(a,b):
    print("a=",a,"b=",b)
cool(5,2)
cool(2,5)
````

````python
# understanding keyword arguments
def print_n(a,b,c):
    print("a",a)
    print("b",b)
    print("c",c)
    print(a,b,c)
    print("calling done")

print_n(55,5,25)
````

````python
def position_demo(a, b, c):
    print(a,b,c)
````

The above code defines a function with the name `position demo`<br>
The  position arguments of the function are added inside the paranthesis of the function<br>
It is to be noted that the same number of arguments have to be used when we are calling the function<br>
The function when called will print the values of a,b and c in the order of the position as provided in the argument

````python
position_demo(2,5,7)
````

Here the value of a=2 ,b=5 and c=7

````python
position_demo(7, 5, 2)
````

Here the value of a=2 ,b=5 and c=7

# 2. Keyword arguments

- Keyword arguments are related to the function calls.


- Here, the caller identifies the arguments by the **parameter name**.   


- Can skip arguments, if you want.


- Can place arguments out of order.


````python
def fn_kwarg(a,b,c):
    print(a,b,c)
    
fn_kwarg(b=10,a=20,c=30)
````

````python
# using positional args
def fn_pos(a,b,c):
    print(a,b,c)
fn_pos(12,6,3) 
# --------------

#using keyword args
def fn_kwarg(a,b,c): 
    print(a,b,c)
fn_kwarg(b=6,a=12,c=3)
````

````python
#setting default value for function arguments
def kwarg_demo(a=2, b=5, c=7):
    print(a,b,c)
    
kwarg_demo()  #not passing any args
````

Here the function name is `kwarg_demo` 
- as we can see in **keyword argument** it is necessary that we assign the variable names to a value (which will act as the default value if nothing is passed while calling of the function)

````python
kwarg_demo(a=2, b=5, c=7)
````

````python
kwarg_demo(b=5, a=2, c=7)
````

````python
#initial value example
kwarg_demo()
````

**Note:**    
- As we can see in the above code, there is no need for a specific position

- In case of keyword argument we just need to specify the values of the variables when we are calling the function

- It does not see the position, it sees the value of the variables

- Here in as the funciton takes a keyword argument it will not give a error

- But rather it will give the default values that we had provided, when we had defined the function

````python
# setting default value both pargs and kwargs

def kwarg(a=0,b=0):
    print(a,b)
    
def pos(d,e):
    print(d,e)
kwarg()
pos()
````

````python
kwarg_demo(15,5) # direct calling
kwarg_demo(b=55,a=67) # giving argument values
````

````python
posKwarg_demo(1,2,4,d=6, f=7, e=0)
````

- **It is to be noted that we use all the position argumnes first and then only the keword arguments***
```
def posKwarg_demo(a,b,c, d = 0, e = 1, f = 3):

```
- Here in the function definition a,b and c are the position arguments
- d,e and f are the keword arguments as the values have been assigned to the arguments

# Functions using Return 

- Functions may return a value to the caller
- Using the keyword- `'return'`

````python
def sum_two_numbers(a, b):
    return a + b
````

````python
c = sum_two_numbers(2,3)
print(c)
````

````python
def add(a, b):
    return a+b
````

````python
def sub(a,b):
    return a-b
````

````python
def mul(a,b):
    return a*b
````

````python
def div(a,b):
    if(b != 0): # try except
        return a/b
    else:
        return "not defined"
````

````python
def algebraic_operations(a,b):
    return( add(a,b), sub(a,b), mul(a,b), div(a,b) )
    
````

# 3. Arbitary arguments
If we do not know how many arguments we are going to use, **we can use the help of arbitary arguments**
- Here we do not have to specify how many arguments we are going to use
Lets see how we can write both position as well as keyword arbitarary arguments

### 1.*args

````python
def arg_demo(*args):
    print(args, type(args))
````

When specifying arbitary arguments we use `*` followed by the argument name
```
def arg_demo(*args):
    print(args, type(args))
```
- Here `arg_demo` is the name of the function
- `*args` is the argument passed inside the function
<p>
- `args` is the name of the function
- since we are using arbitary arguments we put `*` symbol before the name of the argument
- Here the arguments are stored as a tuple 
and hence we can use all the arguments as we use the values in a tuple

````python
arg_demo(2,5,7)
````

````python
def values(*arg):
    print(type(arg))
    for num in arg:
        print(num)
values(1,2,3,4,5,6,7,8)
````

````python
def arg_demo(*args):
    print(args, type(args))
    print(args[0], args[1], args[2])
#     for element in args:
#         print(element)
````

````python
arg_demo(2,5,7,4,5)
````

### 2. kwargs
Here we will see how we can use arbitary keyword arguments

when specifying argument : In case of keyword arguments we use `**` and followed by the argument name

````python
def kwargs_demo(**kwargs):
    print(kwargs, type(kwargs))
````

````python
def student(**arg):
    for keys in arg:
        print(keys,arg[keys])
student(name="sonu",Class=10)
````

````python
kwargs_demo(a = 1, b = 2, c= 3)
````

```
def kwargs_demo(**kwargs):
    print(kwargs, type(kwargs))

kwargs_demo(a = 1, b = 2, c= 3)
```

In the above code we can see that the argument name is `kwargs`<br>
before the name of the code we have `**` symbol specifying that it is a keyword argument<br>
The above function prints the arguments we have passed to the function as well as it also shows the type of the argument<br>
