## 5. Identity Operator
`is` and `is not` are the identity operators both are used to check if two values are located on the same part of the memory.<br> 
> Two variables that are equal does not imply that they are identical
```
is          True if the operands are identical 
is not      True if the operands are not identical
```

````python
# as strings are not mutable they occupy the same memory
a = "name"
b = "name"
print(a is b)
print(a==b)
````

````python
a = 3
b = 3
print(a is b)
````

````python
# as the lists elements can be changed they do not occupy the same memory
a = [1,5,3]
b = [1,5,3]
print(a is b)
````

### Conditional Execution
Checking the condition and doing something with that information<br>

### Understanding if condition
****
#### Syntax
```
if test_condition:
    body of if
```

After `if` next we need to write the test conditon and followed by colon(`:`)

- The body of the if condition is written in the next line
- Remember that the body of the if condition should be intended
    - 4 spaces should be left before the body
<br>
<br>

If the test condition is **True**
- The statement inside the if condition is executed

If the test condition is **False**
- The statement inside the if condition will not executed

**Points To Remember**
- There is no limit to the number of statements that can appear in the body
- There has to be at least one
- In case we need to write a if condition which does not have a statement inside, then we can add pass statement

`Syntax`

```
if test-condition:
    pass 
```
pass  ---->  This does nothing

````python
# testing if statement
x= 10
if x>0: # here test is condition is true 
    print("x is a positive number") # This is body of If condition
print("This program is to understand if condition") # This is not in if condition 
# Question :: what is x>0 -- test condition
````

# Alternative Execution

A second form of if statement is **Alternative Execution**
- here there are two possibilities and the test condition checks which statement has to be executed

`Syntax`

```
if test_condition:
    statements to be executed if
    test condition is True
else:
    statement executed if
    test condtion is False

```

### UNDERSTANDING IF AND ELSE

````python
# Checking if a number is poistive or negative
x = int(input("Enter a number:"))
if x>0:
    print("X is a POSITIVE NUMBER") # Prints if condition is TRUE
else:
    print("x is a negative number") # Prints if condition is FALSE
````
