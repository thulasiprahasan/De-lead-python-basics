```

```

## Understanding while
- If the test condition is true, then the statements inside the while loop is executed
- And after each iteration, the test condition is checked again

### Flow of Execution of while statement
More formally, here is the flow of execution for a while statement: 
1. Evaluate the condition,yielding **True or False**.
2. If the condition is **false**, exit the while statement and continue execution at the next statement.
3. If the condition is **true**, execute the body and then go back to step 1.
- This type of flow is called a **loop** :: because the third step loops back around to the top. 
    - Each time we execute the body of the loop, we call it an **iteration**. 
    - For the below loop, we would say, “It had Ten iterations” which means that the body of of the loop was executed Ten times.

````python

n = 10
while n>0:
    n=n-1
    print(n)
````

## Break Statements

### using Break - for infinite loops
- Infinite loops can be stopped using the break statement   


- Sometimes you don’t know it’s time to end a loop until you get half way through the body.


- In that case you can write an infinite loop on purpose and then use the break statement to jump out of the loop.

````python
# 1 - 5
# use break
a=0
while True:
    a=a+1
    print(a)
    if a==5:
        break
````

````python
# infinite loop :- A loop that runs for ever
count=0
while 1: # while 1 -- 1::TRUE , 0::FALSE
    count = count+1
    print("YES",count)
    if count==10:
        break
print("Ye have put a BREAK!!!")
````

Code Explanation
```
count=0
while 1:
    count = count+1
    print("YES",count)
    if count==10:
        break
print("Ye have put a BREAK!!!")
```
Here we can see that the above while loop is a infinite test condition as it runs for ever
- The test condition is `while 1` which also means `while True`
    - As we know the while loop runs as long as the test condition is true and hence this is a infinite loop
- So now that this is a infinite loop, inorder to stop this infinite loop we use break
- As we can see in the code, a if condition checks when the value of the variable *count is 10* at that time the loop stops (as we are using break)
- As soon as the `break statement` is executed the loop stops

````python
#Real-time example1 for BREAK
# taking inputs from user
# name 
# class
# school
# ------ until user enters done
while True:
    x = input("Enter name of the student:")
    y = input("Enter the class:")
    z = input("Enter the school:")
    a = input("Enter done if you want to stop:")
    if a == "done":
        print("Stopped!!!")
        break
````

````python
#Real-time example2 for BREAK
while True:
    name_1 = input("Enter the name:")
    class_1 = input("Enter Class:")
    school_1 = input("Enter School name:")
    done_1 = input("Enter done if you want to exit or enter any other key:")
    if done_1 == "done":
        break
print("Program Stopped!!!!!")
````

## Continue statement

### Using Continue
- Sometimes you are in an iteration of a loop and want to finish the current iteration and immediately jump to the next iteration.


- In that case you can use the continue


- statement to skip to the next iteration without finishing the body of the loop for the current iteration.


- *Continue with while loop*

````python
#simple example
n=0
while n<5:
    print("*")
    n=n+1 # 1 2 3 4 5
    continue
    print("**")
````

````python
#example2
n=5
while n: # if n==0 - False
    n=n-1
    print(n)
    if n==4:
        continue
    print("*********")
````

#### Method 1 Explanation
```
n=0
while n<11:
    n=n+1
    if n != 5:
        print(n)
```
<h3>Understanding the code</h3>

- Here n is initialised value as 0 
- In line 2 `while n<11` here the test condition is **while n is less than 11** 
    - The loop will run until the n is less than 11 or it can also be said "that is till the test condition is true"
- line 3,4 and 5 are inside the body of the while loop and it will run only when the test condition of the while loop is true
- `n=n+1` initially the value of n is 0, now after this code is executed the value of n becomes 1
- `if n != 5`
    > `!=` This operator is called **NOT EQUAL TO**, it is a combination of two operators `!`(not) and `=`(equal to)
- `if n != 5` the body of the if condition will run only if n is not equal to 5
    - we have `print(n)` inside the body of if
    - which means the it will only print the value of n :- when the value of n is not equal to 5

#### Method 2 Explanation


```
n=0

while n<11:
    n=n+1
    if n == 5:
        continue
    print(n)
```
<h3>Understanding the code</h3>
- The code till if condition is same as before and you can check the above code explanation to understand those lines
- `if n ==5`
    -  If n is equal to 5 then the body of the if condition is printed
    - inside the body of the loop `continue` is used
    - This means that when the if condition becomes **true**
        -  The while loop after that statement wont be executed
        > It will skip the below steps and go to the next iteration
- when the value of n becomes 5 
    - The if condition becomes true
    - The body inside the if condition is executed
    - That means `continue` is executed
    - The code under the continue is not printed 
    - ie. `print(n)` is not printed and it goes to the next iteration
- In the next iteration the value of n is again iterated - `n=n+1` the value of n becomes 6
- So only when the value of n becomes 5 , the if condition becomes **true** and the continue statement is executed
