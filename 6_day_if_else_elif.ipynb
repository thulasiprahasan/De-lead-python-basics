### Elif

The combination of else and if 
- Keyword :: `elif`
*****

Syntax
```
if test_condition_1:
    statement
elif test_condition_2:
    statement
elif test_condition_3:
    statement 
else:
    statement
```

**Points to remember**

- Elif is followed by test condition and ended with colon (:)
- Elif can end with a else or not 
- It is a good practice to add else at the end

As the name suggests Elif is a combination of else and if
Lets first recall else and if
```
if test_condtion:
    statement (executed only if test condition is TRUE)
else :
    statements (Executed if test condtion is False)
```
- If has a test condition is <font color =red>true</font> 
    - The statements inside the if is executed 
- If the test condition is false then the else is executed without checking any test conditions
<br>
Now lets see Elif

```
if test_condtion:
    statement (executed only if test condition is TRUE)
elif test_condition:
    statements (Executed if test condtion is FALSE)
```
    
- If the test condition of if is true we know what happens 
- Now what if the test condition of if becomes FALSE
- The Elif test condition is checked
    - note that in case of else, there was no more of test conditions 
    - If it was else in case of elif the statement inside the else would have printed as soon as test condition became false
    - Here as it is elif (else and if) we can see that when if test condition becomes false, one more test condition is checked and if that test condition becomes True
        - Statement inside the Elif is executed

**Syntax Form:**
```
if condition:
    statements
elif condition:
    statements
elif condition:
    statements
else:
    statements
```

````python
a = int(input("Enter A:"))
b = int(input("Enter B:"))

if a>b:
    print(a," is big")
elif a<b:
    print(b," is big")
else:
    print("Both are equal")
````

### Nested Condtionals
One conditional can be nested with another, this is called as a nested conditional

#### Syntax

```
if test_conditonal_1:
    statement
else:
    if test-conditon_2:
        statement
    else:
        statement
```

Here we can see that inside the first else condition there is if and else condition.<br>
It is not necessary that we need to add a conditional only in else, we can add conditional in if also.

````python
a = 8
if a>99:
    print("3 digit")
else:
    if a<10:
        print("1 digit")
        if a>5:
            print("number is between 6 and 9")
        else:
            print("number is not between 6 and 9")
    else:
        print("2 digit")
````

## Understanding Try and Except

**Syntax**
```
try :
    Statements 
except:
    Statements to be executed if error occurs in try
```

- here the condition we want to test is written in the try,
    - make sure to note the intendation
- If a error occurs when the try condition is executed then the statements inside the `except` is executed

Try and except is very useful to check if we have any errors in program and if error occues how to execute it otherwise.

````python
try:
    number = int(input("Enter a number:"))
    print("This is just checking")
    print(number + 10)
    
except:
    print("Wrong input")
````

````python
try :
    input_user = int(input("Enter a number:"))
    print("Number",input_user)
except:
    print("\t !!!Wrong Input!!!")
````

````python
#division by zero
a,b = int(input("Enter first number: ")),int(input("Enter second number: "))
print('quotient=',a//b)
````

````python
#division by zero error can be excluded
a,b = int(input("Enter first number: ")),int(input("Enter second number: "))
try:
    print('quotient=',a//b)
except:
    print("invalid inputs")
````
