# For Loop

- It can iterate over the items of any sequence such as list, string, tuple, dictionary etc
- It has three parts:
    - iterating variable
    - sequence (list, string, tuple, dictionary etc)
    - body of the loop (set of statments need to be executed repeatedly)

- **Syntax:**
```
for iter_variable in sequence:
            statements
```

````python
# Question 2 : numbers = [22,55,44,66,88,77]
# print the sum of the numbers in the list numbers using for loop
# hint 
# 22 + 55 = 77
# 77 + 44 = 121
# 121 + 66 = 187
# 187 + 88 = 275
# 275 + 77 = 352

sum_a = 0
count = 0
numbers = [22,55,44,66,88,77] # sum = 22+55+44+66+88+77 = 352
for a in numbers:
    count =count +1
    print("value of a",a)
    print("loop =",count)
    sum_a = sum_a + a
    print("sum=",sum_a)
print("Final ",sum_a)
````

## Break

#### Print the elements till 35(excluded) , in the given  list below
```
num = [1,22,55,85,66,35,45,85,55]
```

````python
num = [1,22,55,85,66,35,45,85,55]
for n in num:
    if n==35:
        break
    print(n)
````

````python
num = [1,22,55,85,66,35,45,85,55]
for n in num:
    if n==num[5]:
        break
    print(n)
````

````python
# Use Break Statement
# Print the elements till 35(exclueded) , in the given below list
num = [1,22,55,85,66,35,45,85,55]
for number in num:
    print(number)
    if(number==66):
        break
````

## Nested For loops

- Loop inside another loop

````python
# for loop inside a for loop
adj = ["red", "big", "tasty"]
fruits = ["apple", "banana", "cherry"]

for x in adj:
    for y in fruits:
        print(x, y)
````

## Range

Range with a single argument <br>
`range(argument)`
- here range will print numbers from 0 till the value passed in the argument, in incrementation of 1


````python
# using only single argument in range
# using range with For Loop
for num in range(5,11):
    print(num)
````

````python
# using only single argument in range
# using range with For Loop
for num in range(0,11,2):
    print(num)
````

```
for num in range(10):
    print(num)
```
Lets see how this code works
- Num will have values from 0 till 9(as range(10))
- after each iteration the value of num will be incremented with 1
    - loop 1 :: num =0
    - loop 2 :: num =1
    ...... and so on 
    - loop 10 :: num =9
- Hence the above code will print numbers from 0 till 9

for a in range(5):
    for b in range(a):
        print(b)

# Enumerate

- enumerate() method adds a counter to an iterable and returns it in a form of enumerating object.


**Syntax:**
`enumerate(iterable, start=0)` <br>
```
    Where iterable is any sequence that can be iterate
          start is the index in which couter need to be started
```

````python
for i,j in enumerate([1,2,3]):
    print(i,j)
````

````python
numbers = [2,3,5,7,11,13,17,19]
for item in enumerate(numbers):
    print(item)
````

````python
numbers = [2,3,5,7,11,13,17,19]
for index,item in enumerate(numbers):
    print(index,item)
````

````python
numbers = [2,3,5,7,11,13,17,19]
for index,item in enumerate(numbers,10):
    print(index,item)
````

## Else with Loop

- In python, an else statement can be associated with a loop statement


- It is an optional feature


- If the else statement is used with a for loop, the else statement is executed when the loop has exhausted iterating the list.


- If the else statement is used with a while loop, the else statement is executed when the condition becomes `False`.

````python
# Else in for loop
names = ['Arun','AnitHa','Bincy','Fayis','Nikhil']

for name in names:
    print(name)
else:
    print('Name list had ended!')
````

````python
n = 10

while(n>=0):
    print(n)
    n = n-1
else:
   print("Reached negative numbers")
````
