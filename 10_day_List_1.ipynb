## Indexing

- Each element inside the list is given a specific position

- Each element inside the list is given a specific position called as Index

Which means that
- First element is given the index value - 0
- Second element is given the index value - 1 and so on ..

- As the values are indexed it is possible
    - To change a value at a specific index (as lists are mutable)
    - Add elements using index
    - Get the value of elements in a specific location

## Transversal through a list

### Transversal using while loop

````python
new_list = ["areej","akarsh","amrutha","shahan","arjun","akshaya","raya","adarsh"]
length = len(new_list)
print("length:",length)
a=0
while a<8:
    print(new_list[a]) 
    a+=1
````

````python
# While Loop
# using while loop print all the elements in a list
random = ["names","numbers",1,2,3,"float",225.6,True,False]
x=0
while x<len(random):
    print(random[x])
    x=x+1
````

### Traversal of list elements using for loop

**Syntax:**
```
for item in list_name:
    print(item)
```
- Here item is just an identifier you can give any name
- In each iteration of the for loop `item` carries one element of the list.
- Iteration will occur `n-times`, where `n` is the number of elements in the list.
- Iteration will take elements from the list in the order they appeared in the list. i.e from index position `0` to `-1`.

**Note:** In this case we need not to worry about the index.

````python
new_list = ["areej","akarsh","amrutha","shahan","arjun","akshaya","raya","adarsh"]
for name in new_list:
    print(name)
````

````python
#To print index along with the item
new_list = ["areej","akarsh","amrutha","shahan","arjun","akshaya","raya","adarsh"]
a=0
for name in new_list:
    a=a+1
    print(a,name)
````

````python
random = ["names","numbers",1,2,3,"float",225.6,True,False]
print(random)
print("------")
for a in random:
    print(a)
````

## Changing the values inside a list

**Note:** As lists are mutable, it is possible to change the values inside a list


- This can be achieved using indices and assignment operator


- No extra method is needed for updating a list element

**Syntax:**
```

```
- Here index is the index of the element which we want to altrered.

````python
marks = [65,76,94,72,80]
# We want to change the mark 72 into 75
print("Before: ",marks)
marks[3] = 75
# marks[-2] = 75
print("After: ",marks)
````

## List Slicing and negative indexing

**Syntax:**
```
sub_list = list_name[start_index:end_index:step]
```

````python
random = ["names","numbers",1,2,3,"float",225.6,True,False]

# names - 0
# numbers - 1
# 1 - 2
# 2 - 3
# 3 - 4
# float - 5
# 225.6 - 6
# True - 7
# False - 8

print(random[:])
print(random[5:8])
print(random[:5]) # 5 excluding 
print(random[5:]) # 5 including
````

````python
random = ["names","numbers",1,2,3,"float",225.6,True,False]
# names - 0
# numbers - 1
# 1 - 2
# 2 - 3
# 3 - 4
# float - 5
# 225.6 - 6
# True - 7
# False - 8
print(random[:8])
print(random[:8:1])
print(random[:8:2])
````
